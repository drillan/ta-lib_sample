```python
import pandas as pd
import matplotlib.pyplot as plt
import talib
from talib import MA_Type
from talib import abstract
```

```python
df = pd.read_csv('./data/nikkei_stock_average_daily_jp.csv', parse_dates=['date'], index_col='date')
```

### SMA

```python
sma = talib.SMA(df['close'])
```

```python
fig = plt.figure(figsize=(10, 4))
ax = fig.add_subplot(1, 1, 1)
ax.plot(df['close'])
ax.plot(sma)
plt.show()
```

### BBANDS

```python
upper, middle, lower = talib.BBANDS(df['close'], matype=MA_Type.T3)
```

TEMA、Triple Exponential Moving Average
> 三重指数移動平均 https://www.tradingtechnologies.com/help/x-study-ja/technical-indicator-definitions/tema/

```python
fig = plt.figure(figsize=(10, 4))
ax = fig.add_subplot(1, 1, 1)
ax.plot(df['close'])
ax.plot(upper)
ax.plot(middle)
ax.plot(lower)
plt.show()
```

### Abstract API

#### SMA

```python
sma = abstract.SMA(df, timeperiod=30)
```

```python
fig = plt.figure(figsize=(10, 4))
ax = fig.add_subplot(1, 1, 1)
ax.plot(df['close'])
ax.plot(sma)
plt.show()
```

#### Stochastic

```python
stoch_df = abstract.STOCH(df, 5, 3, 0, 3, 0)
```

```python
fig = plt.figure(figsize=(10, 4))
ax = fig.add_subplot(1, 1, 1)
ax.plot(stoch_df['slowk'].tail(100))
ax.plot(stoch_df['slowd'].tail(100))
plt.show()
```
